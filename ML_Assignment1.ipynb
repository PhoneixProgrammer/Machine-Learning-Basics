```python
df = pd.DataFrame({'US spending on science': {'1999': 18000000000,
  '2000': 18500000000,
  '2001': 19500000000,
  '2002': 20500000000,
  '2003': 20800000000,
  '2004': 23000000000,
  '2005': 23500000000,
  '2006': 23500000000,
  '2007': 25500000000,
  '2008': 27500000000,
  '2009': 29500000000},
 'Hanging sucides': {'1999':5400,
  '2000': 5600,
  '2001': 6000,
  '2002': 6400,
  '2003': 6600,
  '2004': 7300,
  '2005': 7200,
  '2006': 7400,
  '2007': 8000,
  '2008': 8500,
  '2009': 9000}})
```

```python
uss = df['US spending on science']
hs = df['Hanging sucides']
years = df.index

plt.figure(figsize=(20,5))
plt.title('US spending on science, space and technology \n correlates with\n Sucides by hanging strangulation and suffocation', fontsize=20)

ax1 = plt.gca()
line1, = ax1.plot(years, uss,'-d',c='r',label='US spending on science')
ax1.set_ylabel('US spending on science', c='r',fontsize=16)
ax1.tick_params(axis='y', labelcolor='r')

ax2 = ax1.twinx()
line2, = ax2.plot(years, hs, '-o',c='black',label="Hanging sucides")
ax2.set_ylabel("Hanging sucides",fontsize=16)
ax2.set_yticks([4000,6000,8000,10000])
ax2.set_yticklabels(['4000 sucides','6000 sucides', '8000 sucides', '100000 sucides'], rotation='horizontal', fontsize=14)

ax3=ax1.twiny()
line3,=ax3.plot(years, uss,'-d',c='r')

ax1.set_yticks([15000000000,20000000000,25000000000,30000000000])
ax1.set_yticklabels(['$15 billion','$20 billion', '$25 billion', '$30 billion'], rotation='horizontal', fontsize=14)

ax1.legend(bbox_to_anchor=(0.4,-0.3), loc="lower center",fontsize=16,frameon=False)
ax2.legend(bbox_to_anchor=(0.6, -0.3), loc="lower center",fontsize=16,frameon=False)

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)

ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_visible(False)

ax1.grid(axis='y')

plt.tick_params(axis='x', which='major', labelsize=14)
ax1.tick_params(axis='x', which='major', labelsize=14)
```

```python
sample_data=pd.read_csv('https://raw.githubusercontent.com/pratoolbharti/NIU/master/Applied_Machine_Learning_S20/Homework-1/mpg.csv')
```

```python
sample_data
```

```python
fwd_data = sample_data.loc[sample_data['drv'] == 'f']
rwd_data = sample_data.loc[sample_data['drv'] == 'r']
awd_data = sample_data.loc[sample_data['drv'] == '4']
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,8),constrained_layout=True)
fig.suptitle('Partial Transparency and jittering',fontsize=30)
ax[0,0].title.set_text('Initial plot')
ax[0,0].scatter(fwd_data.displ,fwd_data.cty,color='orange',label='FWD')
ax[0,0].scatter(rwd_data.displ,rwd_data.cty,color='blue',label='RWD')
ax[0,0].scatter(awd_data.displ,awd_data.cty,color='black',label='4WD')
ax[0,0].legend()
ax[0,0].set_xlabel("displacement (l)")
ax[0,0].set_ylabel("fuel economy (mpg)")

ax[0,1].title.set_text('Transparency')
ax[0,1].scatter(fwd_data.displ,fwd_data.cty,color='orange',label='FWD',alpha=0.5)
ax[0,1].scatter(rwd_data.displ,rwd_data.cty,color='blue',label='RWD',alpha=0.5)
ax[0,1].scatter(awd_data.displ,awd_data.cty,color='black',label='4WD',alpha=0.5)
ax[0,1].legend()
ax[0,1].set_xlabel("displacement (l)")
ax[0,1].set_ylabel("fuel economy (mpg)")

ax[1,0].title.set_text('Jittering')
fwd_displ_jitter = fwd_data.displ + np.random.normal(0, 0.1, size=len(fwd_data))
fwd_cty_jitter = fwd_data.cty + np.random.normal(0, 0.1, size=len(fwd_data))
ax[1,0].scatter(fwd_displ_jitter,fwd_cty_jitter,color='orange',label='FWD',alpha=0.5)
rwd_displ_jitter = rwd_data.displ + np.random.normal(0, 0.1, size=len(rwd_data))
rwd_cty_jitter = rwd_data.cty + np.random.normal(0, 0.1, size=len(rwd_data))
ax[1,0].scatter(rwd_displ_jitter,rwd_cty_jitter,color='blue',label='RWD',alpha=0.5)
awd_displ_jitter = awd_data.displ + np.random.normal(0, 0.1, size=len(awd_data))
awd_cty_jitter = awd_data.cty + np.random.normal(0, 0.1, size=len(awd_data))
ax[1,0].scatter(awd_displ_jitter,awd_cty_jitter,color='black',label='4WD',alpha=0.5)
ax[1,0].legend()
ax[1,0].set_xlabel("displacement (l)")
ax[1,0].set_ylabel("fuel economy (mpg)")

ax[1,1].title.set_text('Over Jittering')
fwd_displ_jitter = fwd_data.displ + np.random.normal(0, 0.9, size=len(fwd_data))
fwd_cty_jitter = fwd_data.cty + np.random.normal(0, 0.9, size=len(fwd_data))
ax[1,1].scatter(fwd_displ_jitter,fwd_cty_jitter,color='orange',label='FWD',alpha=0.5)
rwd_displ_jitter = rwd_data.displ + np.random.normal(0, 0.9, size=len(rwd_data))
rwd_cty_jitter = rwd_data.cty + np.random.normal(0, 0.9, size=len(rwd_data))
ax[1,1].scatter(rwd_displ_jitter,rwd_cty_jitter,color='blue',label='RWD',alpha=0.5)
awd_displ_jitter = awd_data.displ + np.random.normal(0, 0.9, size=len(awd_data))
awd_cty_jitter = awd_data.cty + np.random.normal(0, 0.9, size=len(awd_data))
ax[1,1].scatter(awd_displ_jitter,awd_cty_jitter,color='black',label='4WD',alpha=0.5)
ax[1,1].legend()
ax[1,1].set_xlabel("displacement (l)")
ax[1,1].set_ylabel("fuel economy (mpg)")

plt.show()
```
